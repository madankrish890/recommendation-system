```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

# Load data from CSV file
#df = pd.read_csv('ratings.csv')

# Create pivot table to get user-item matrix
user_item_matrix = df.pivot_table(index='userId', columns='movieId', values='rating')

# Fill NaN values with 0
user_item_matrix.fillna(0, inplace=True)

# Calculate cosine similarity between users
user_similarity = cosine_similarity(user_item_matrix)

# Define function to get top recommendations for a user
def get_user_recommendations(user_id, num_recommendations=10):
    # Get index of user
    user_index = user_item_matrix.index.get_loc(user_id)
    # Calculate similarity scores between user and all other users
    sim_scores = list(enumerate(user_similarity[user_index]))
    # Sort scores in descending order
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    # Get IDs of top similar users
    similar_users = [i[0] for i in sim_scores[1:]]
    # Get all items rated by similar users, but not by the current user
    recommendations = user_item_matrix.iloc[similar_users].sum(axis=0)
    recommendations = recommendations[recommendations == 0]
    # Sort items by rating and get top recommendations
    recommendations = recommendations.sort_values(ascending=False)[:num_recommendations]
    # Get original titles of recommended movies
    recommended_movies = df.loc[df['movieId'].isin(recommendations.index)]['original_title'].unique()
    return recommended_movies
```

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

# Load data from CSV file
#df = pd.read_csv('ratings.csv')

# Create pivot table to get item-user matrix
item_user_matrix = df.pivot_table(index='movieId', columns='userId', values='rating')

# Fill NaN values with 0
item_user_matrix.fillna(0, inplace=True)

# Calculate cosine similarity between items
item_similarity = cosine_similarity(item_user_matrix.T)

# Define function to get top recommendations for an item
def get_item_recommendations(item_id, num_recommendations=10):
    # Get index of item
    item_index = item_user_matrix.index.get_loc(item_id)
    # Calculate similarity scores between item and all other items
    sim_scores = list(enumerate(item_similarity[item_index]))
    # Sort scores in descending order
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    # Get IDs of top similar items
    similar_items = [i[0] for i in sim_scores[1:]]
    # Get all users who rated the similar items and their ratings
    ratings = item_user_matrix.iloc[similar_items]
    # Calculate average rating for each user
    avg_ratings = ratings.apply(lambda x: x[x!=0].mean(), axis=0)
    # Sort users by average rating and get top recommendations
    recommendations = avg_ratings.sort_values(ascending=False)[:num_recommendations]
    # Get original titles of recommended movies
    recommended_movies = df.loc[df['userId'].isin(recommendations.index)]['original_title'].unique()
    return recommended_movies[:10]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df['Text']=df['overview'].fillna('')+''+df['genres_name'].fillna('').apply(lambda x:' '.join(x))
```

```python
import string
import re
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from bs4 import BeautifulSoup
from nltk.tokenize import word_tokenize
```

```python
stop_words=stopwords.words('english')
lemmatizer=WordNetLemmatizer()
```

```python
def preprocess(text):
    # remove HTML tags and URLs
    text = re.sub(r'http\S+', '', text)
    text = BeautifulSoup(text, "html.parser").get_text()
    # remove non-alphabetical characters and digits
    text = re.sub('[^a-zA-Z\s]', '', text)
    # convert to lowercase
    text = text.lower()
    # tokenize the text
    tokens = word_tokenize(text)
    # remove stopwords
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in tokens if word not in stop_words]
    # lemmatize the tokens
    lemmatizer = WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(word) for word in filtered_tokens]
    # join the tokens back into a string
    text = ' '.join(lemmatized_tokens)
    return text
```

```python
df['Text1']=df['Text'].apply(preprocess)
```

```python
tfidf=TfidfVectorizer()
tf_matrix=tfidf.fit_transform(df['Text1'])
```

```python
simi=cosine_similarity(tf_matrix,tf_matrix)
```

```python
def recommendations(title,simi,df):
    
    idx=df[df['original_title']==title].index[0]
    
    sim_cos=list(enumerate(simi(idx)))
    
    sim_cos=sorted(sim_cos,key=lambda x:x[0],reverse=True)
    
    sim_scores=sim_cos[1:11]
    
    movie_indices=[i[0] for i in sim_scores]
    
    return df['original_title'].iloc[movie_indices]
```

```python
recommendations('Heat',simi=simi,df=df)
```
